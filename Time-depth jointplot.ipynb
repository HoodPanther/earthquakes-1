```python
#libraries
import pandas as pd #датафреймы
import numpy as np #массивы, линейная алгебра
import pylab as pl #графики
import math
import matplotlib as mpl #графики
%matplotlib inline
%pylab inline
import datetime
import seaborn as sns #красивые графики!
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# encoding=utf8  
import sys  
```

```python
def plot(path):
    
    df=pd.read_csv(path, sep = ' ', index_col=False, names = ['time','lat','lon','depth','Mb','Mw'])
    t = []
    depth = []
    for time in df['time'].unique():
        df_time = df[df['time']==time]
        t.append(time)
        depth.append(df_time.loc[df_time['Mb'].idxmax()]['depth'])
    t = pd.Series(t)
    depth = pd.Series(depth)
    t = t.diff()
    t = t[1:]
    depth = depth[1:]
    df = pd.DataFrame()
    df['time'] = t
    df['depth'] = depth
    print 'Файл:', path
    print 'Количество событий:', len(df)
    print 'Совместное распределение глубины и промежутков времени'

    
    sns.jointplot(x='time', y='depth', data=df, size=6,kind='kde');
    sns.plt.title(u'Совместное распределение глубины и промежутков времени', loc='right', fontsize=13, y=1.24)
    
    plt.show()
```

```python
for file in ['Japan.txt', 'Japan_C.txt']:
    plot(file)
```

```python
for file in ['M_Atl_R.txt','M_Atl_R_C.txt']:
    plot(file)
```
